```python
# Retrieve the predictions from the job
full_predictions = job.get_predictions()

# Print the full predictions to check their contents
print("Full predictions:", full_predictions)

# Initialize an empty list to store all transcriptions
all_transcriptions = []

# Iterate over each prediction in the list
for prediction in full_predictions:
    # Extract the transcriptions from the prediction
    transcriptions = prediction.get('transcriptions', [])
    # Add the transcriptions to the list of all transcriptions
    all_transcriptions.extend(transcriptions)

# Print the all transcriptions to check their contents
print("All transcriptions:", all_transcriptions)

# Print each transcription on a new line
for i, transcription in enumerate(all_transcriptions, 1):
    print(f"Transcription {i}: {transcription['transcript']}")
```
